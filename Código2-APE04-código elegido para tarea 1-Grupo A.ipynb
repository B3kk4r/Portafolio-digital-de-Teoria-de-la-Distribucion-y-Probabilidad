```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint, norm, binom, uniform
import seaborn as sns

 # Configuracion de estilo
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams["font.size"] = 11
print("Entorno configurado correctamente")
```

##Para este ejercicio, he seleccionado la Opción C: Distribución Binomial. Esta es una variable aleatoria discreta excelente para modelar situaciones con un número fijo de ensayos donde solo hay dos resultados posibles.
##Contexto del Problema:
###Para garantizar la operatividad del Laboratorio de Computación, se monitorea la estabilidad de los equipos. La sala principal cuenta con 15 computadoras a disposición de los estudiantes. Según los registros de mantenimiento del sistema, la probabilidad de que un equipo individual experimente un conflicto de actualizaciones y necesite ser reiniciado durante la jornada académica es del 20%.


```python
# Definición de la distribución

# Parametros de la distribucion binomial
n = 15  # numero de ensayos (computadoras)
p = 0.2  # probabilidad de exito (reinicio)

# Crear la distribucion
dist_binomial = binom(n, p)

# Rango de valores para graficar (valores discretos)
x = np.arange(0, n + 1)

# PMF
pmf_binomial = dist_binomial.pmf(x)

# Visualizacion
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.stem(x, pmf_binomial, basefmt=" ", linefmt='b-', markerfmt='bo', label=f'B({n}, {p})')
plt.axvline(n*p, color='red', linestyle='--', label=f'Media = {n*p}')
plt.xlabel('x (Número de computadoras reiniciadas)')
plt.ylabel('p(x)')
plt.title('PMF: Variable Aleatoria Discreta (Binomial)')
plt.legend()
plt.grid(alpha=0.3)
plt.xticks(x)

# CDF
cdf_binomial = dist_binomial.cdf(x)
plt.subplot(1, 2, 2)
plt.step(x, cdf_binomial, where='post', color='red', linewidth=2)
plt.scatter(x, cdf_binomial, color='red', zorder=5)
plt.xlabel('x')
plt.ylabel('F(x) = P(X <= x)')
plt.title('CDF: Distribucion Binomial')
plt.grid(alpha=0.3)
plt.xticks(x)
plt.ylim(0, 1.1)

plt.tight_layout()
plt.show()
```

```python
# Cálculo de probabilidades

# a) P(X = 3), probabilidad de que la variable sea exactamente 3
prob_igual_3 = dist_binomial.pmf(3)
print(f"P(X = 3) = {prob_igual_3:.4f} ({prob_igual_3*100:.2f}%)")

# b) P(X <= 2), probabilidad de que la variable este en un rango menor o igual a 2
prob_menor_2 = dist_binomial.cdf(2)
print(f"P(X <= 2) = {prob_menor_2:.4f} ({prob_menor_2*100:.2f}%)")

# c) P(X > 4), probabilidad de que la variable sea mayor a 4
prob_mayor_4 = 1 - dist_binomial.cdf(4)
print(f"P(X > 4) = {prob_mayor_4:.4f} ({prob_mayor_4*100:.2f}%)")

# d) Percentil 95, el número de reinicios máximo dado el 95% de las veces
p95 = dist_binomial.ppf(0.95)
print(f"\nPercentil 95: {p95:.0f}")
print(f"El 95% de las veces, los reinicios son menores o iguales a {p95:.0f}")
```

##Interpretación de Resultados


###1.   P(X = 3) = 0.2501 (25.01%): Existe aproximadamente un 25% de probabilidad de que exactamente 3 de las 15 computadoras del laboratorio necesiten ser reiniciadas durante el día. Según la gráfica de PMF, este es uno de los escenarios más probables.
###2.   P(X <= 2) = 0.3980 (39.80%): Hay casi un 40% de probabilidad de que el número de computadoras que necesiten reiniciarse sea bajo (ninguna, 1, o 2 computadoras). Esto nos lo confirma la gráfica CDF, donde vemos que el escalón sube rápido al inicio.
###3.   P(X > 4) = 0.1642 (16.42%): Es poco probable (solo un 16.42%) que se necesite reiniciar a más de 4 computadoras en un solo día. Esto indica que un evento masivo de reinicios simultáneos es raro bajo estas condiciones.
