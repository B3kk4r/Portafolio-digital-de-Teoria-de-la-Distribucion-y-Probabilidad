```python
#Se importan las librerías a usar
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns
```

```python
# Parámetros para la creación de la binomial
ensayos = 100          # número de ensayos
prob_independiente = 0.5         # probabilidad de cara o cruz

# Definir distribución binomial por medio de los parametros anteriores
dist = binom(ensayos, prob_independiente)

# Valores que toma x en 100 ensayos
x = np.arange(0, ensayos+1)

# Creación de PMF y CDF
pmf = dist.pmf(x)
cdf = dist.cdf(x)

plt.figure(figsize=(12,5))
# Graficación de la PMF
plt.subplot(1,2,1)
plt.bar(x, pmf, color="skyblue", edgecolor="black")
plt.title("PMF - Distribución Binomial de moneda")
plt.xlabel("Número de caras en {} lanzamientos".format(ensayos))
plt.ylabel("Probabilidad para cada punto")

# Graficación de la CDF
plt.subplot(1,2,2)
plt.step(x, cdf, where="post", color="red")
plt.title("CDF - Distribución Binomial de la moneda")
plt.xlabel("Número de caras en {} lanzamientos".format(ensayos))
plt.ylabel("Probabilidad acumulada")

plt.tight_layout()
plt.show()

print("Esto muestra la distribución de valores cuando a una moneda se expone a 100 ensayos, cada uno independiente del otro y con la misma probabilidad \n")

# Calcular las probabilidades específicas
# Para cuando P(X = 50)
p_exacto_10 = dist.pmf(50)

# Para P(30 <= X <= 40)
prob_entre_30_40 = dist.cdf(40)-dist.cdf(29)

# Para P(X > 60)
p_mas_60 = 1 - dist.cdf(60)

print(f"P(X = 50) = {p_exacto_10:.4f}")
print("Esta es la probabilidad de que x tome el valor de 50 \n")

print(f"P(30 <= X <= 40)   = {prob_entre_30_40:.4f}")
print("Esta es la probabilidad de que x tome un valor ubicado entre 30 y 40, incluyendo los a ambos \n")

print(f"P(X > 60) = {p_mas_60:.4f}")
print("Esta es la probabilidad de que x tome un valor mayor a 60 \n")

```
